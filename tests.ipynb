```python
train_df = pd.read_csv('data/wikitext-103/train.csv')
```

```python
#train_df.head()
def get_doc_count(train_df):
    return train_df.shape[0]

def get_doc(train_df, i):
    return train_df.iat[i, 0]
```

```python
import re
import html

re1 = re.compile(r'  +')
def fixup(x):
    x = x.replace('#39;', "'").replace('amp;', '&').replace('#146;', "'").replace(
        'nbsp;', ' ').replace('#36;', '$').replace('\\n', "").replace('\n', "").replace('quot;', "'").replace(
        '<br />', "\n").replace('\\"', '"').replace('<unk>','u_n').replace(' @.@ ','.').replace(
        ' @-@ ','-').replace('\\', ' \\ ')
    return re1.sub(' ', html.unescape(x)).encode('ascii',errors='ignore').decode("ascii", errors="ignore")
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

doc = nlp(fixup(get_doc(train_df, 0)))
```

```python
def tokenize(doc):
    tokens = []
    for token in doc:
        #if(token.text.isalnum()):
        tokens.append(token.text.lower())
    return tokens

tokenize(doc)
```
